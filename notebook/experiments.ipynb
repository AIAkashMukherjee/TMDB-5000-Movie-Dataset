```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
credit_df=pd.read_csv('../data/tmdb_5000_credits.csv')
movies_df=pd.read_csv('../data/tmdb_5000_movies.csv')
```

```python
movies_df = movies_df.merge(credit_df,on='title')
movies_df.head(2)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies_df.info()
```

```python
int_column=movies_df.select_dtypes(exclude=['object']).columns
int_column
```

```python
str_column=movies_df.select_dtypes(include=['object']).columns
str_column
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df['cast']=movies_df['cast'].apply(convert)
movies_df['crew']=movies_df['crew'].apply(convert)
```

```python
movies_df.head()
```

```python
def create_tags(x):
    return ' '.join(x['genres']) + ' ' + ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + ' '.join(x['crew'])
```

```python
movies_df['tags']=movies_df.apply(create_tags, axis=1)
movies_df
```

```python
movies_df=movies_df.drop(columns=['overview','budget','homepage','production_companies','revenue','runtime','spoken_languages','status','id','genres','keywords','production_countries','vote_count','original_language','release_date','title','crew','vote_average','popularity','tagline','cast'])
```

```python
movies_df.head(4)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['tags'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
movies = movies_df.reset_index()
indices = pd.Series(movies.index, index=movies['original_title']).drop_duplicates()
```

```python
indices
```

```python
# Function to get movie recommendations
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return movies['original_title'].iloc[movie_indices]
```

```python
# Test the recommendation system
print(get_recommendations('Avatar'))
```
